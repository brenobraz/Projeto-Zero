```python
import scipy as sp
import matplotlib.pyplot as plt
from ipywidgets import interact
%matplotlib inline
```

```python
k = 1
wn = 2*2*sp.pi
wf = 2.1*2*sp.pi
f0 = 1
qsi = 0.15
t = sp.linspace(0,2,1000)
omega = wf/wn
```

```python
def f(wf):
    x = (f0/k)*((1-omega**2)*sp.cos(wf*t) + 2*qsi*omega*sp.sin(wf*t))/((1-omega**2)**2 + (2*qsi*omega)**2)
    plt.plot(t,x)
```

```python
interact(f,wf=5)
```
